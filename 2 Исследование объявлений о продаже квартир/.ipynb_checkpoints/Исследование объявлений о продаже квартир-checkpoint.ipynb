# Исследование объявлений о продаже квартир

**Цель проекта:**

Целью данного проекта является проведение исследовательского анализа данных по объявлениям о продаже квартир в Санкт-Петербурге и соседних населенных пунктах. Мы стремимся определить факторы, влияющие на цену недвижимости, а также выявить закономерности в продажах квартир, такие как среднее время продажи и зависимость цены от различных параметров.

**Описание данных:**
    
- *airports_nearest* — расстояние до ближайшего аэропорта в метрах (м)
- *balcony* — число балконов
- *ceiling_height* — высота потолков (м)
- *cityCenters_nearest* — расстояние до центра города (м)
- *days_exposition* — сколько дней было размещено объявление (от публикации до снятия)
- *first_day_exposition* — дата публикации
- *floor* — этаж
- *floors_total* — всего этажей в доме
- *is_apartment* — апартаменты (булев тип)
- *kitchen_area* — площадь кухни в квадратных метрах (м²)
- *last_price* — цена на момент снятия с публикации
- *living_area* — жилая площадь в квадратных метрах (м²)
- *locality_name* — название населённого пункта
- *open_plan* — свободная планировка (булев тип)
- *parks_around3000* — число парков в радиусе 3 км
- *parks_nearest* — расстояние до ближайшего парка (м)
- *ponds_around3000* — число водоёмов в радиусе 3 км
- *ponds_nearest* — расстояние до ближайшего водоёма (м)
- *rooms* — число комнат
- *studio* — квартира-студия (булев тип)
- *total_area* — общая площадь квартиры в квадратных метрах (м²)
- *total_images* — число фотографий квартиры в объявлении

# Предобработка данных<a id="2"></a> 

Напишем несколько функций для обработки данных

```python
def count_empty_share(x):     # возвращает долю пропусков в столбце x
    return f'Доля пропусков в столбце {x} составляет  {round((data[x].isna().sum()/len(data))*100, 4)}%'

def find_ub_lb(x):     # возвращает границы нехарактерных значений
    lb = data[x].describe()[4] - ((data[x].describe()[6] - data[x].describe()[4]) * 1.5)
    ub = data[x].describe()[6] + ((data[x].describe()[6] - data[x].describe()[4]) * 1.5)
    return (lb, ub)

# следующие функции убирают неявные дубли в названии
def change_name(x):     # заменяет букву ё в слове поселок
    if isinstance(x, str):
        e = list(x.split())
        if 'посёлок' in e:
            e[e.index('посёлок')]=e[e.index('посёлок')].replace('ё', 'е')
            return ' '.join(e)
    return x

def change_name_2(x):     # меняет дубли (поселок городского типа/поселок)
    if isinstance(x, str):
        e = list(x.split())
        if e[0] + ' ' + e[-1] in data['locality_name'].unique() and len(e) > 2:
            return e[0] + ' ' + e[-1]
    return x
```

**first_day_exposition**

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
print('Начало датасета:', data['first_day_exposition'].min())
print('Конец датасета:', data['first_day_exposition'].max())
```

**ceiling_height**

```python
count_empty_share('ceiling_height')
```

```python
# заменим опечатки в высоте потолка (будем считать, если высота указана выше 20 - то клиенты ошиблись на 1 порядок)
data['ceiling_height'] = data['ceiling_height'].apply(lambda x: x/10 if x>=20 else x)
# заменим пропуски на медианы в зависимости от этажности дома, для домов с отсутствием этажности заменим на общую медиану
data.loc[(data['ceiling_height'].isna()) 
             & (data['floors_total'].isna()), 'ceiling_height'] = data['ceiling_height'].describe()[5]
for floor in data['floors_total'].unique():
    data.loc[(data['ceiling_height'].isna()) 
             & (data['floors_total'] == floor), 
             'ceiling_height'] = data.query('floors_total == @floor')['ceiling_height'].describe()[5]
data=data[~data['ceiling_height'].isna()].copy()
count_empty_share('ceiling_height')
```

```python
# границы нехарактерных значений
find_ub_lb('ceiling_height')
```

```python
data.boxplot('ceiling_height', vert=False)
plt.xlim(1, 5)
plt.title('Диаграмма размаха')
plt.xlabel('Высота потолков, м')
```

### **floors_total**

```python
count_empty_share('floors_total')
```

```python
#Заменим пропуски в столбце 'floors_total' на значения столбца 'floor' и приведем столбец к типу int
data.loc[data['floors_total'].isna(), 'floors_total'] = data['floor']
data = data.astype({'floors_total': 'int'})
count_empty_share('floors_total')
```

```python
# границы нехарактерных значений
find_ub_lb('floors_total')
```

```python
data.boxplot('floors_total', vert=False)
plt.title('Диаграмма размаха')
plt.xlabel('Этажи')
```

**living_area**

```python
count_empty_share('living_area')
```

```python
# заменим пустые значения на долю от общей площади в зависимости от этажности дома
share_living = data.pivot_table(index='floors_total', values=['living_area', 'total_area'])
share_living['share_area'] = share_living['living_area']/share_living['total_area']
for floor in share_living.index:
    data.loc[(data['living_area'].isna()) 
             & (data['floor'] == floor), 'living_area'] = data['total_area'] * share_living.loc[floor, 'share_area']
count_empty_share('living_area')
```

```python
# границы нехарактерных значений
find_ub_lb('living_area')
```

```python
data.boxplot('living_area', vert=False)
plt.title('Диаграмма размаха')
plt.xlabel('Жилая площадь, м2')
```

```python
#Заменим пропуски в категории на 'unknown'
data.loc[data['is_apartment'].isna(), 'is_apartment'] = 'unknown'
data['is_apartment'].value_counts()
```

**kitchen_area**

```python
count_empty_share('kitchen_area')
```

```python
# поступим с пропусками в этом столбце аналогично столбцу living_area
# заменим пустые значения на долю от общей площади в зависимости от этажности дома
share_kitchen = data.pivot_table(index='floors_total', values=['kitchen_area', 'total_area'])
share_kitchen['share_area'] = share_kitchen['kitchen_area']/share_kitchen['total_area']
for floor in share_kitchen.index:
    data.loc[(data['kitchen_area'].isna()) 
             & (data['floor'] == floor), 'kitchen_area'] = data['total_area'] * share_kitchen.loc[floor, 'share_area']
count_empty_share('kitchen_area')
```

#### Для дальнейшего анализа исключим из датафрейма выбросы

```python
data['ceiling_height_good'] = ((data['ceiling_height'] >= find_ub_lb('ceiling_height')[0]) & 
                               (data['ceiling_height'] <= find_ub_lb('ceiling_height')[1]))
data['floors_total_good'] = ((data['floors_total'] >= find_ub_lb('floors_total')[0]) & 
                               (data['floors_total'] <= find_ub_lb('floors_total')[1]))
data['living_area_good'] = ((data['living_area'] >= find_ub_lb('living_area')[0]) & 
                               (data['living_area'] <= find_ub_lb('living_area')[1]))
data['kitchen_area_good'] = ((data['kitchen_area'] >= find_ub_lb('kitchen_area')[0]) & 
                               (data['kitchen_area'] <= find_ub_lb('kitchen_area')[1]))
data['total_area_good'] = ((data['total_area'] >= find_ub_lb('total_area')[0]) & 
                               (data['total_area'] <= find_ub_lb('total_area')[1]))
data['last_price_good'] = ((data['last_price'] >= find_ub_lb('last_price')[0]) & 
                               (data['last_price'] <= find_ub_lb('last_price')[1]))
good_data = data.query('ceiling_height_good and floors_total_good and \
                       living_area_good and kitchen_area_good and total_area_good and last_price_good').copy()
len(good_data)/len(data)*100
```

### Вывод:

Проведена предобработка данных, для дальнейшего анализа исключены 14.8% данных из датафрейма.

Для дальнейшего анализа добавим в таблицу новые столбцы с ценой за м2, днем публикации объявления, месяцем публикации объявления, годом публикации объявления, типом этажа квартиры, а так же с расстоянием до центра города в км.

```python
# добавим столбец с ценой за м2
good_data['price_one_meter'] = good_data.apply(lambda x: int(x[1]/x[2]), axis=1)
# добавим столбец с днем недели публикации объявления
good_data['day_of_week_exposition'] = good_data['first_day_exposition'].dt.dayofweek
# добавим столбец с месяцем публикации объявления
good_data['month_exposition'] = good_data['first_day_exposition'].dt.month
# добавим столбец с годом публикации объявления
good_data['year_exposition'] = good_data['first_day_exposition'].dt.year
# добавим столбец с типом этажа квартиры (значения — «первый», «последний», «другой»)
def categorization_floor(x):
    if x[8]==1:
        return 'первый'
    if x[8]==x[6] and x[8]!=1:
        return 'последний'
    return 'другой'
good_data['category_floors'] = good_data.apply(categorization_floor, axis=1)
# добавим столбец с расстоянием до центра города в километрах
good_data['city_centers_km'] = good_data['cityCenters_nearest'].apply(lambda x: round(x/1000) if not np.isnan(x) else x)
```
