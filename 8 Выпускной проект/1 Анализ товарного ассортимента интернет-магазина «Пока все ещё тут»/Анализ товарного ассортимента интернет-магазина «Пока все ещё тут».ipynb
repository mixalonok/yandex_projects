Напишем функцию для построения плотности распределения

```python
def plot_distribution(df, clmn, xl='Количество', yl='Плотность', title='График распределения плотности', wdt=10, hgt=6):
    """Функция получает на вход параметры для построения графика распределения плотности

    df - датафрейм
    clmn - столбец, по которому строится график
    xl, yl, title, wdt, hgt - параметры графика
    """
    fig, ax = plt.subplots(figsize=(wdt,hgt))
    plt.xlabel(xl)
    plt.ylabel(yl)
    plt.title(title)
    sns.kdeplot(data=df[clmn])
    ax.xaxis.grid(True)
    ax.yaxis.grid(True)
    plt.show()
```

Построим график распределения плотности количества товаров

## Анализ товарного ассортимента<a id="eda1"></a> 

Для дальнейшего анализа необходимо изучить названия товаров и разделить товары на группы. Для этого приведём все слова в наименовании товара к лемме.

```python
lemmatizer = Mystem()
def lemmatize_text(text):
    """Функция получает на вход переменную и возвращает лемму
    """
    lemmas = lemmatizer.lemmatize(text)
    return ' '.join(lemmas).strip()
data_clean['lemmas'] = data_clean['product'].apply(lemmatize_text)
data_clean.sample(2)
```

Выделим первые два слова, определяющие категорию товара, исключим предлоги итеративным путём (просматривая вывод функции и корректируя её).

```python
def extract_first_two_words(text):
    """Функция получает на вход текст и выводит первые два слова без предлогов, характеристик и цифр
    """
    words = text.split()
    words = [word for word in words if not word.lower() in ['в', 'на', 'за', 'и', 'для', '-', 'd',\
             'с', 'c', 'по', 'см', '1л', '(', 'h', 'под', 'со', 'мл', '350х180', '350х150', 'о', 'мм', '150х120',\
              'л', '40х40', 'из', 'Е4', '214х150', 'E4']\
             and not word.isdigit()]
    return ' '.join(words[:2])
```

```python
data_clean['lemmas_two_words'] = data_clean['lemmas'].apply(extract_first_two_words)
print(*data_clean.lemmas_two_words.unique(), sep='///')
```

```python
all_text = ' '.join(data_clean['lemmas_two_words'].tolist())
words = re.findall(r'\w+', all_text.lower())
word_counts = Counter(words)
popular_words = word_counts.most_common()
for word, count in popular_words:
    print(f'{word}: {count}')
```

### Определяем основной ассортимент<a id="eda2"></a> 

Проведём объединенный ABC-XYZ анализ. Для этого выполним следующие шаги:

- 1) Расчет общей выручки и количества для каждого продукта:


- 2) Сортировка продуктов по убыванию стоимости:


- 3) Расчет кумулятивной стоимости и процентного соотношения:

    - Вычислим кумулятивную стоимость для каждого продукта, суммируя стоимость продуктов в порядке убывания.
    - Вычислим процентное соотношение кумулятивной стоимости к общей стоимости всех продуктов.
    
    
- 4) Классификация продуктов на категории ABC:

    - Категория A: Продукты, составляющие 80% от общей стоимости продуктов.
    - Категория B: Продукты, составляющие следующие 15% от общей стоимости продуктов.
    - Категория C: Продукты, составляющие оставшиеся 5% от общей стоимости продуктов.


- 5) Расчет кумулятивного количества и процентного соотношения:
    
    - Рассчитаем среднемесячную выручку
    - Рассчитаем стандартное отклонение и коэффициент вариации (КВ)
    

- 6) Классификация продуктов на категории XYZ:

    - Категория Х — стабильный спрос, объем продаж от месяца к месяцу почти не изменяются: КВ < 0,1.
    - Категория Y — условно стабильный спрос, т. е. объемы продаж меняются, но без резких скачков: 0,25 > КВ > 0,1.
    - Категория Z — спрос нестабилен, покупатели приобретают товар от случая к случаю: КВ > 0,25.

Рассчитаем общую выручку и количество для каждого продукта и отсортируем по убыванию стоимости

```python
product_summary = data_clean.groupby('product').agg({'earn': 'sum', 'quantity': 'sum'})
product_summary_sorted = product_summary.sort_values('earn', ascending=False)
product_summary_sorted.head()
```

Рассчитаем кумулятивную выручку и её процентное соотношение к общей выручке

```python
product_summary_sorted['cumulative_earn'] = product_summary_sorted['earn'].cumsum()
product_summary_sorted['percentage_earn'] = product_summary_sorted['cumulative_earn'] \
/ product_summary_sorted['earn'].sum()
product_summary_sorted.head()
```

Классифицируем продукты на категории A, B, C и добавим категории в наш датасет

```python
product_summary_sorted['category_ABC'] = 'C'
product_summary_sorted.loc[product_summary_sorted['percentage_earn'] <= 0.8, 'category_ABC'] = 'A'
product_summary_sorted.loc[(product_summary_sorted['percentage_earn'] \
                            > 0.8) & (product_summary_sorted['percentage_earn'] <= 0.95), 'category_ABC'] = 'B'
data_clean = data_clean.merge(product_summary_sorted['category_ABC'], on='product', how='left')
data_clean.sample(5)
```

Посчитаем количество товара в каждой категории

```python
data_clean.drop_duplicates('product').groupby('category_ABC')['date'].count()
```

Классифицируем продукты на категории XYZ, для этого рассчитаем среднемесячную выручку.

```python
xyz = pd.pivot_table(data_clean, index='product', columns='year_month', values='earn', aggfunc='sum')
xyz['mean_earn'] = xyz.sum(axis=1)/16
xyz = xyz.fillna(0)
xyz.sample(5)
```

Посчитаем стандартное отклонение

```python
std_dev = xyz.iloc[:, :-1].std(axis=1)
xyz_std = xyz.copy()
xyz_std['std_dev'] = std_dev
xyz_std.head()
```

Найдём коффициент вариации

```python
xyz_std['coef_var'] = xyz_std['std_dev'] / xyz_std['mean_earn']
xyz_std.sample(5)
```

Объявим функцию для классификации товаров по категориям XYZ

```python
def xyz_code(x):
    '''Функция получает на вход число и выводит категорию'''
    if x <= 0.1:
        return 'X'
    if x <= 0.25:
        return 'Y'
    return 'Z'
```

Применим функцию и посчитаем количество товара в каждой категории

```python
xyz_std['category_XYZ'] = xyz_std['coef_var'].apply(xyz_code)
xyz_std.groupby('category_XYZ')['mean_earn'].count()
```

## Формулировка и проверка статестических гипотез <a id="eda3"></a> 

**Гипотеза 1**: Существует связь между категорией товара и его поппулярностью.

**Описание**: Предположим, что разные категории товаров (основной ассортимент A и дополнительный B) имеют различную популярность.

- H0: Нет разницы в популярности между товарами категории A и категории B.
- H1: Товар A популярнее товара B.

```python
product_A = data_clean[data_clean['category_ABC'] == 'A']['quantity']
product_B = data_clean[data_clean['category_ABC'] == 'B']['quantity']
t_stat, p_value = ttest_ind(product_A, product_B)
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: Товар категории A популярнее товара категори B")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```

**Гипотеза 2**: Существует связь между категорией товара и его ценой.


**Описание**: Предположим, что разные категории товаров имеют различные ценовые диапазоны. Например, товары из категории "Растения, муляжи и искусственные цветы" могут иметь более высокие цены по сравнению с товарами из категории "Текстиль для дома". Проверим, есть ли статистически значимая разница в средних ценах между разными категориями товаров.

- H0: Средняя цена товаров не зависит от категории.
- H1: Средняя цена товаров различается в зависимости от категории.

```python
category_A_prices = data_clean[data_clean["category_ABC"] == "A"]["price"]
category_B_prices = data_clean[data_clean["category_ABC"] == "B"]["price"]
t_statistic, p_value = stats.ttest_ind(category_A_prices, category_B_prices, equal_var=False)
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: Средняя цена товаров различается в зависимости от категории")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```

**Гипотеза 3**: Количество покупок зависит от дня недели.


**Описание**: Предположим, что количество покупок может различаться в зависимости от дня недели. Например, в выходные дни может быть больше покупок, чем в будние дни. Проверим, существует ли статистически значимая разница в среднем количестве покупок в разные дни недели.

- H0: Количество покупок не различается в зависимости от дня недели.
- H1: Количество покупок зависит от дня недели.

```python
weekday_counts = data_clean.groupby('week')["quantity"].sum()
weekday_counts = data_clean[data_clean["date"].dt.weekday < 5]["quantity"]
weekend_counts = data_clean[data_clean["date"].dt.weekday >= 5]["quantity"]
t_statistic, p_value = stats.ttest_ind(weekday_counts, weekend_counts, equal_var=False)
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: количество покупок не различается в зависимости от дня недели")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```

**Гипотеза 4**: Существует сезонность в покупках определенных товаров.


**Описание**: Предположим, что некоторые товары имеют сезонную популярность и покупаются чаще в определенные месяцы или времена года. Например, товары для праздничного украшения могут быть более популярными в декабре. Проверим, есть ли статистически значимая разница в средних количествах покупок для разных месяцев.

- H0: Количество покупок не различается в зависимости от сезонности.
- H1: Количество покупок зависит от сезонности.

```python
monthly_counts = data_clean.groupby('month')["quantity"].sum()
# Применим однофакторный дисперсионный анализ (ANOVA) для сравнения средних значений количества покупок между месяцами
month_groups = [group["quantity"] for _, group in data_clean.groupby('month')]
f_statistic, p_value = stats.f_oneway(*month_groups)
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: количество покупок не различается в зависимости от сезонности")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```
