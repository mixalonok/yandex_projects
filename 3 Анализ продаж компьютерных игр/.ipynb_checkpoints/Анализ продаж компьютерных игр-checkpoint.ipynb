# Анализ продаж компьютерных игр

**Цель проекта:**

Целью данного проекта является проведение анализа данных о продажах компьютерных игр для определения закономерностей, влияющих на успешность игр. Это позволит интернет-магазину "Стримчик" спланировать рекламные кампании и сделать ставку на потенциально прибыльные продукты.

**Описание данных:**
- *Name* — название игры
- *Platform* — платформа
- *Year_of_Release* — год выпуска
- *Genre* — жанр игры
- *NA_sales* — продажи в Северной Америке (миллионы проданных копий)
- *EU_sales* — продажи в Европе (миллионы проданных копий)
- *JP_sales* — продажи в Японии (миллионы проданных копий)
- *Other_sales* — продажи в других странах (миллионы проданных копий)
- *Critic_Score* — оценка критиков (максимум 100)
- *User_Score* — оценка пользователей (максимум 10)
- *Rating* — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Подготовка данных<a id="1"></a> 

Приведём названия столбцов к нижнему регистру

```python
data.columns = map(lambda x: x.lower(), data.columns)
```

```python
data[data['name'].isna()]
```

Имя и жанр пропущены всего у 2х строк, что составляет 0,0119% от общего датасета. Удалим их.

```python
data = data.dropna(subset = ['name'])
data.name.isna().sum()
```

Проверим на уникальность название платформ, жанра и игр

```python
data['platform'] = data['platform'].astype('str')
sorted(data['platform'].unique())
```

```python
data['name'] = data['name'].astype('str')
sorted(data['name'].unique())
```

```python
data['genre'] = data['genre'].astype('str')
sorted(data['genre'].unique())
```

Приведем год релиза к типу данных int

```python
data['year_of_release'] = data['year_of_release'].fillna(0)
data['year_of_release'] = data['year_of_release'].astype('int')
print(f'Датасет представлен c {data.year_of_release.min()} по {data.year_of_release.max()}')
```

Заменим пустые года на 0, в дальнейшем попытаемся их обработать.

```python
data[data['name'].isna()]
```

Обработаем пропуски в столбце с годом релиза

```python
display(data.loc[data['year_of_release']==0]['year_of_release'].count())
data.loc[data['year_of_release']==0].head(5)
```

```python
data.loc[data['name']=='FIFA Soccer 2004'].head(5)
```

Заметим, что в датасете одна и та же игра может быть на разных платформах. Подтянем год релиза, там где это возможно, с других платформ

```python
for i in data.query('year_of_release==0').index:
    name_of_game = data.loc[i, 'name']
    data.loc[i, 'year_of_release'] = data.query('name == @name_of_game')['year_of_release'].max()
display(data.loc[data['year_of_release']==0]['year_of_release'].count())
```

Осталось 146 пропусков, удалим эти строки.

```python
data = data.query('year_of_release != 0')
```

```python
data['user_score'].unique()
```

В столбце с пользовательской оценкой есть значение tbd - скорее всего, это "to be determined", что означает, что оценка ещё не определена, соответственно можем заменить это значение на nan.

```python
data.loc[data['user_score']=='tbd', 'user_score'] = np.nan
```

Приведём оценки пользователей к 100 бальной шкале и типу инт, а так же оценки критиков к типу инт

```python
data['user_score'] = data['user_score'].apply(lambda x: x if pd.isnull(x) else float(x)*10)
data = data.astype({'user_score': 'Int64', 'critic_score': 'Int64'}, errors='ignore')
data.head(5)
```

Посчитаем суммарные продажи во всех регионах и сведём их в один столбец

```python
data['all_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data.head(5)
```

```python
pd.pivot_table(data, index='year_of_release', values='rating', aggfunc='count').plot()
plt.title('График рейтинга')
plt.xlabel("Год выпуска")
plt.ylabel("Количество рейтинговых оценок");
```

Вывод по предобработке данных:

Были обработаны пропуски, заменены значения с датой релиза игры (датой релиза на других платформах), Строки с пропущенной датой релиза были удалены. Согласно вики ESRB - основана в 1994 году, соответственно и рейтинг начали присваивать с 1994 года. Скорее всего, пропуски обусловлены этим фактом. То же самое и с оценками. Игровые журналы, которые публиковали оценки критиков и пользователей появились позднее чем сами игры.

#### Посмотрим изменение продаж по платформам

```python
all_sales_platform = pd.pivot_table(data, index='platform', values='all_sales', aggfunc='sum')
sales_platform = pd.\
pivot_table(data, index='platform', columns='year_of_release', values='all_sales', aggfunc='sum')\
.join(all_sales_platform).query('all_sales > 100').sort_values('all_sales', ascending=False)

for column in sales_platform.columns:
    if sales_platform[column].sum() == 0:
        sales_platform = sales_platform.drop([column], axis=1)
sales_platform = sales_platform.drop(['all_sales'], axis=1)
sales_platform = sales_platform.T
sales_platform
```

Выберем самые поппулярные платформы (до 100 млн продаж) и оценим изменение продаж по годам, а так же среднее "время жизни" платформы

```python
bar_width = 0.4
plt.figure(figsize=(15, 8))
current_xticks = plt.xticks()[0]
plt.xticks(current_xticks, rotation=90)
bar_positions = np.arange(len(sales_platform.index))
for i, platform in enumerate(sales_platform.columns):
    if i == 0:
        plt.bar(bar_positions, 
                sales_platform[platform], 
                bar_width, 
                label=platform);
    else:
        plt.bar(bar_positions, 
                sales_platform[platform], 
                bar_width, 
                bottom=sales_platform.iloc[:, :i].sum(axis=1), 
                label=platform);
plt.title('Продажи по платформам в разрезе лет')
plt.xlabel('Год')
plt.ylabel('Продажи, млн')
plt.xticks(bar_positions + bar_width, sales_platform.index)
plt.legend()
plt.show()
```

```python
average_life_time = {platform:[sales_platform[platform].last_valid_index()-sales_platform[platform].first_valid_index(), 
                               sales_platform[platform].first_valid_index(), 
                               sales_platform[platform].last_valid_index()] 
                     for i, platform in enumerate(sales_platform.columns)}
top_platform_info = pd.DataFrame.from_dict(average_life_time, orient='index', columns=['life_time', 'start', 'finish'])
print(f'Срендее время жизни платформы, достигшей 100млн продаж: {top_platform_info.life_time.mean():0.1f} лет')
top_platform_info
```

```python
#Проверим коэфициент корреляции оценок с продажами у платформ, по которым строк с оценками больше 100
def corr2type(platforma):
    if len(data_fly[(data_fly.platform==platforma) & ~(data_fly.critic_score.isna())]) < 100:
        return
    critic_coef = data_fly.loc[(data_fly.platform==platforma) & ~(data_fly.critic_score.isna()), 'critic_score']
    critic_sales = data_fly.loc[(data_fly.platform==platforma) & ~(data_fly.critic_score.isna()), 'all_sales']
    user_coef = data_fly.loc[(data_fly.platform==platforma) & ~(data_fly.user_score.isna()), 'user_score']
    user_sales = data_fly.loc[(data_fly.platform==platforma) & ~(data_fly.user_score.isna()), 'all_sales']
    print(f'{platforma}:')
    print('Корреляция оценки критиков и продаж:', 
          round(critic_coef.astype('float').corr(critic_sales), 2))
    print('Корреляция оценки игроков и продаж:', 
          round(user_coef.astype('float').corr(user_sales), 2))
    print()

for i in data_fly.platform.unique():
    corr2type(i)
```

Корреляция практически на всех платформах слабая, но корреляция с оценкой игроков намного слабее чем корреляция с оценкой критиков. Это может объясняться тем, что игроки больше доверяют критикам и игровым экспертам чем таким же пользователям. Так же, наглядно заметно что в зоне с высокими оценками критиков - больше 70 баллов - больше игр, которые выстреливают по продажам

Предположение о том что оценка критиков больше коррелирует с продажами подтвердилась и на других платформах.

## Проверка гипотез <a id="5"></a>

Сформулируем две гипотезы:
   - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
   - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
   
И проверим их.

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Тогда нулевая и альтернативная гипотезы будут звучать так:
- H0: Средние пользовательские рейтинги платформ Xbox One и PC равны
- H1: Средние пользовательские рейтинги платформ Xbox One и PC отличаются

Для проверки, возьмем две выборки с оценками платформ Xbox One и PC.
Критический уровень стат значимости зададим 5%

```python
xo = list(data_fly.loc[(data_fly.platform == 'XOne') & ~(data_fly.user_score.isna()), 'user_score'])
pc = list(data_fly.loc[(data_fly.platform == 'PC') & ~(data_fly.user_score.isna()), 'user_score'])
alpha = 0.05
results = st.ttest_ind(xo, pc)
print('p-уровень значимости:', results.pvalue)
if results.pvalue <alpha:
    print('Нулевая гипотеза не подтвердилась')
else:
    print('Не удалось опровергнуть нулевую гипотезу')
```

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Тогда нулевая и альтернативная гипотезы будут звучать так:
- H0: Средние пользовательские рейтинги жанров Action и Sports равны
- H1: Средние пользовательские рейтинги жанров Action и Sports отличаются

Для проверки, возьмем две выборки с оценками жанров Action и Sports.
Критический уровень стат значимости зададим 5%

```python
action = list(data_fly.loc[(data_fly.genre == 'Action') & ~(data_fly.user_score.isna()), 'user_score'])
sports = list(data_fly.loc[(data_fly.genre == 'Sports') & ~(data_fly.user_score.isna()), 'user_score'])
alpha = 0.05
results = st.ttest_ind(action, sports)
print('p-уровень значимости:', results.pvalue)
if results.pvalue <alpha:
    print('Нулевая гипотеза не подтвердилась')
else:
    print('Не удалось опровергнуть нулевую гипотезу')
```
